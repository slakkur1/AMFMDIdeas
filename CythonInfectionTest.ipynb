```python
%%cython
cimport numpy as np
ctypedef np.int_t DTYPE_t
ctypedef np.float64_t DTYPE_d

# Define expontential function from C math lib.
cdef extern from "math.h":
    double exp(double arg)

def exp_func(arg):
    return exp(arg)

def infection_process_cython(np.ndarray[DTYPE_t, ndim = 1] status, np.ndarray[DTYPE_d, ndim=2] MaxRate, np.ndarray[DTYPE_t, ndim = 1] grid, np.ndarray[DTYPE_d, ndim = 1] Transmiss, np.ndarray[DTYPE_d, ndim = 1] Suscept, np.ndarray[DTYPE_t, ndim = 1] first_in_grid, np.ndarray[DTYPE_t, ndim = 1] last_in_grid, int period_latent, np.ndarray[DTYPE_t, ndim = 1] Num, np.ndarray[DTYPE_d, ndim = 2] K):
    # Type all variables used
    cdef int slen = status.shape[0]
    cdef np.ndarray[DTYPE_t, ndim = 1] INF = np.where( (status > (1 + period_latent) ))[0], Event = np.zeros(slen)
    cdef np.ndarray[DTYPE_t, ndim = 1] IGrids, m
    cdef np.ndarray[DTYPE_d] R, trans, maxr, rate, MaxProb, rng
    cdef int NI = INF.shape[0], s, ind1, INFi, ii, j, n, M, leng
    cdef float PAB, Q, P#, rated, MaxProbd
    
    IGrids = grid[INF] # find the grid cells of infected farm locations
    ########################################
    # SUSCEPTIBLE TO INFECTIOUS TRANSITION #
    ########################################
    # Loop through all infectious farms
    #print "Starting loop through all infectious farms (infection_process function)"
    #print "here 1"
    for ii in range(NI):
        # Assign INFi the infected farm's index
        INFi = INF[ii]
        ################################# replacement loop, used for speed testing.  
        #R = np.random.rand(Ngrids)
        #for iii in range(Ngrids):
        #    rated = -Transmiss[INFi]*Num[iii]*MaxRate[IGrids[ii],iii]
        #    MaxProbd = 1 - exp_func(rated)
        #    if (R[iii] < MaxProbd):
        #        m[iii] = 1
        #m = np.where(m == 1)[0]
        #################################
        #print "here 2"
        trans = -Transmiss[INFi]*Num
        #print "here 3"
        maxr = MaxRate[IGrids[ii],:]
        #print "here 4"
        # Elementwise multiplication
        rate = np.multiply(trans, maxr)
        #print "here 5"
        MaxProb = 1 - np.exp(rate)
        #print "here 6"
        rng = np.random.rand(len(MaxProb))
        m = np.where( MaxProb - rng > 0)[0]
        # these grids need further consideration
        for n in range(len(m)):
            s = 1
            M = m[n] # Loop thru grids where an infection event may have occurred.  
            #print "here 7"
            PAB = 1 - exp_func(-Transmiss[INFi]*MaxRate[IGrids[ii],M])
            #print "here 7.1"
            if (PAB == 1):
                # Calculate the infection probability for each farm in the susceptible grid
                leng = last_in_grid[M] - first_in_grid[M] + 1
                R = np.random.rand(leng)
                #print "here 7.2"
                for j in range(leng):
                    ind1 = first_in_grid[M] + j
                    #print "Here 7.3", j
                    Q = 1 - exp_func(-Transmiss[INFi]*Suscept[ind1]*K[INFi, ind1])
                    if ((R[j] < Q) & (status[ind1] == 0)):
                        #print "This is an infection:", ind1
                        Event[ind1] = 1
                ###############################
                #print "here 8"
                #ind = np.arange(start = first_in_grid[M], stop = last_in_grid[M]+1, dtype = DTYPE)
                #print "here 9"
                #tmp = np.exp(-Transmiss[INFi]*Suscept[ind]*K[INFi, ind])
                #print "Finished tmp"
                #Q = np.add(-tmp, 1)
                #print "here 10"
                #test = (np.random.rand(len(Q))-Q < 0) & (status[ind]==0)
                #print "here 11"
                #Event[ind[test]] = 1
            else:
                # Loop through all susceptible farms in the grids where an infection event occurred.  
                #print "here 8.1"
                R = np.random.rand(Num[M])
                #print "here 9.1"
                for j in range(Num[M]):
                    #print "here 10.1"
                    ind1 = first_in_grid[M] + j
                    #print "here 11.1"
                    P = 1 - s*(1 - PAB)**(Num[M] - j)
                    #print "here 12.1"
                    if (R[j] < (PAB / P)):
                        #print "here 14.1"
                        s = 0
                        #print "here 15.1"
                        Q = 1 - exp_func(-Transmiss[INFi]*Suscept[ind1]*K[INFi, ind1])
                        #print "here 16.1"
                        if (R[j] < Q/P) & (status[ind1] == 0):
                            #print "here 17.1"
                            Event[ind1] = 1
    # Evolve the infection process of those farms which have been exposed and already infectious ones.  
    status[status > 0] += 1
    status = status + Event
    #print "Ended loop through all infectious farms"
    return(status, Event)

# np.ndarray[int, ndim = 1] # np.ndarray[np.int_t, ndim = 1]
def Update_Grid_Probabilities_Vectorized_cython(gridz, DTYPE_t grid_n_x, DTYPE_t grid_n_y, np.ndarray[DTYPE_d, ndim = 1] Suscept_, np.ndarray[DTYPE_d, ndim = 2] KDist2all):
    
    """
    wjp11@psu.edu
    W. Probert, 2014
    """
    # Type all variables used
    cdef np.ndarray[DTYPE_t, ndim = 1] Num, first_, last_
    cdef np.ndarray[DTYPE_d, ndim = 1] max_sus
    cdef np.ndarray[DTYPE_d, ndim = 2] MaxRate1
    cdef int index
    print "1"
    # Find the farm indices per grid square
    farms_per_grid = map(lambda index: np.where(gridz == index)[0], range(grid_n_x*grid_n_y) )
    print "2"
    # Maximum susceptibility of a feedlot per grid square.  
    max_sus = np.asarray([np.max(np.append(Suscept_[farms_per_grid[index]], 0)) for index in range(grid_n_x*grid_n_y)])
    print "3"
    # Number of farms per grid square
    Num = np.array(map(len, farms_per_grid))
    print "4"
    # WHAT ARE THESE USED FOR?  
    # First farm in grid, -1 if the list is empty
    
    # HAVE TO USE A DIFFERENT WAY TO CONSTRUCT THIS HERE... CYTHON DOESN'T LIKE NAN in INTEGER ARRAYS... 
    first_ = np.array(map(lambda index: np.nanmin(np.append(index, np.nan)), farms_per_grid))
    print "5"
    #first_ = np.asarray(first_)
    first_[np.isnan(first_)] = -1
    #first_ = first_.astype(int)
    print "6"
    # Last farm in grid, -2 if the list is empty
    last_ = np.array(map(lambda index: np.max(np.append(index, -2)), farms_per_grid))
    print "7"
    #last_ = np.asarray(last_)
    #last_ = last_.astype(int)
    
    MaxRate1 = np.multiply(KDist2all, max_sus) # this takes advantage of numpy broadcasting
    print "8"
    # All grids with no farms should have infinite rate of input infection
    # All grids with no farms should have infinite rate of susceptibility
    MaxRate1[Num == 0] = np.inf
    MaxRate1[:, Num == 0] = np.inf
    print "9"
    
    # Set rate from one grid to itself to infinity.  
    np.fill_diagonal(MaxRate1, np.inf)
    print "10"
    return(MaxRate1, Num, first_, last_, max_sus)

#ctypedef np.float64_DTYPE_d
#I wanted to try and run cython block by using "%%cython" but it doesnt seem to like the combination of cython and numpy commands
```
